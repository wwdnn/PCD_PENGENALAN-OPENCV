```python
from PIL import Image
import cv2  
from IPython.display import display

# Memuat gambar
img = cv2.imread('../GAMBAR/Gambar3_PCD.jpeg')

# Mengkonversi gambar dari BGR ke RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Menampilkan gambar
display(Image.fromarray(img))

# Menampilkan color channel
r, g, b = img[0,0]
print("Red: ", r)
print("Green: ", g)
print("Blue: ", b)

# newline
print()

# Menampilkan color channel
r, g, b = cv2.split(img)
print("Red: ", r)
print("Green: ", g)
print("Blue: ", b)
```

```python
# Menampilkan gambar grayscale dengan color channel merah
pil_032 = Image.fromarray(r)
display(pil_032)
```

```python
# Menampilkan gambar grayscale dengan color channel hijau
pil_032 = Image.fromarray(g)
display(pil_032)
```

```python
#  Menampilkan gambar grayscale dengan color channel biru
pil_032 = Image.fromarray(b)
display(pil_032)
```

```python
# Mengganti nilai color channel r, g, b dengan nilai b, g, r
merged = cv2.merge([b, g, r])

# Menampilkan gambar dengan color channel b, g, r
pil_032 = Image.fromarray(merged)
display(pil_032)
```

```python
# Menginputkan nilai color channel dengan nilai r, g, b
merged = cv2.merge([r, g, b])

# Menampilkan gambar dengan color channel r, g, b
pil_032 = Image.fromarray(merged)
display(pil_032)
```

```python
import numpy as np
import copy

r_dup = copy.deepcopy(r)
g_dup = copy.deepcopy(g)
b_dup = copy.deepcopy(b)
r_dup[:] = 0
g_dup[:] = 0
b_dup[:] = 0

merged = cv2.merge([r, g_dup, b_dup])
pil_032 = Image.fromarray(merged)
display(pil_032, "Ketika Seluruh Nilai Channel g dan b Menjadi 0")

merged = cv2.merge([r_dup, g, b_dup])
pil_032 = Image.fromarray(merged)
display(pil_032, "Ketika Seluruh Nilai Channel r dan b Menjadi 0")

merged = cv2.merge([r_dup, g_dup, b])
pil_032 = Image.fromarray(merged)
display(pil_032, "Ketika Seluruh Nilai Channel r dan g Menjadi 0")

merged = cv2.merge([r_dup, g_dup, b_dup])
pil_032 = Image.fromarray(merged)
display(pil_032, "Ketika Seluruh Nilai Channel r, g, dan b Menjadi 0")
```
